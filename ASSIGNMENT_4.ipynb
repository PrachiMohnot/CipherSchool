###### * Objective

To perform EDA(Exploratory Data Analysis) on the given Adult Dataset.

```python
#Load the data to adult dataframe

missing_values=['?']
adult=pd.read_csv("adult.csv",na_values=missing_values)
```

```python
# Data Overview

adult.head()
```

```python
# Number of Rows and columns in adult dataset

actual_data=adult.shape
act_rows=actual_data[0]
actual_data
```

```python
# Brief Statistical description of the dataset

adult.describe(include='all')
```

######  Conclusion:

- There are 48842 number of data enteries and 15 attributes
- Most people are 'White' 'Males' having 'private' workclass with income '<=50K'
- Mean age is 38.64
- People on average work 40 hours per week



```python
# Concise summary of dataset

adult.info()
```

```python
# Removing 'fnlwgt' from adult dataframe

adult=adult.drop('fnlwgt', axis=1)
adult
```

```python
# Dropping rows having missing values

adult=adult.dropna(subset=['workclass', 'occupation','native-country'])
```

```python
# Final summary of modified adult dataset


adult.describe()
```

```python
# To calculate missing value percentage:

mdfied_rows=adult.shape[0]
fractn=(act_rows-mdfied_rows)/act_rows
percntg=np.round(100*fractn,2)
print("Enteries removed: ",(act_rows-mdfied_rows))
print("Lost percentage is: ",percntg,"%")
```

######  Conclusion:

- There are total 2799 missing values in 'workclass', 2809 in 'occupation' and 857 in 'native-country' column.
- Column 'fnlwgt' is dropped since it is of no use.
- While removing missing values, total 3620 enteries have been removed, i.e. 7.41% data has been lost.
- Mean and Median of all the columns have small differences, except 'capital-gain' and 'capital-loss'.

# Bivariate Analysis

### * To discover the relationships between numerical columns.

- Considering our target variable is 'income'.

```python
# Pairplot between 3 numericals cols: age, educational number, hours per week.

sns.pairplot(adult, vars=['age','educational-num','hours-per-week'],corner=True,kind='scatter')
plt.show()
```

<b>Observatons:</b>
<br><b> age v/s educational number:</b>
- Here, no significant relation can be observed between age and education number.
<br><b> age v/s hours-per-week:</b>
- People with age above 75, tend to work less than 60 hours per week.
<br><b> educational number v/s hours-per-week:</b>
- no significant relation can be observed between hours per week and education number.


```python
# Plots of capital gain with age, educational number, hours per week.

sns.jointplot(adult['capital-gain'],adult['age'],alpha=0.3)
plt.show()
sns.jointplot(adult['capital-gain'],adult['educational-num'],alpha=0.3)
plt.show()
sns.jointplot(adult['capital-gain'],adult['hours-per-week'],alpha=0.3)
plt.show()
```

<b>Observtions:</b>
-  No significant relation cab be found b/w age and capital gain.
- maximum capital(appx.100,000) is gained by people having higher educational number.
- maximum capital(appx.100,000) is gained by people working between 35-60 hours per week.

```python
# plots of capital loss with age, educational number, hours per week.

sns.jointplot(adult['age'],adult['capital-loss'],alpha=0.3)
plt.show()
sns.jointplot(adult['educational-num'],adult['capital-loss'],alpha=0.3)
plt.show()
sns.jointplot(adult['hours-per-week'],adult['capital-loss'],alpha=0.3)
plt.show()
```

<b>Observtions:</b>
- Mode of capital loss is 1,500-3,000.
- people aged less than 60 faced loss more than above 60 years old.
- people having higher educational number faced loss more than lower ed-num.
- people working 35-60 hours per week faced loss.
