```python
# Q6 - Display a breif overview of all the columns data type and missing values?

cols=df.columns
types=df.dtypes
miss_value=df.isnull().sum()
missing_percent=round((100*(df.isnull().sum()/len(df.index))),2)
data={}
for i in range(len(cols)):
    data[cols[i]]=[types[i],miss_value[i],missing_percent[i]]
new_df1=pd.DataFrame(data,["Data Types","Missing Values","Missing Value Percent"])
new_df1

```

```python
# Q9 - Display the percentage of working men and women in the dataframe?

print("Total Enteries: ",len(df.index))

data=round((df['gender'].value_counts()*100)/len(df.index),2)

print("percentage of males: ",data[0],"%\npercentage of females: ",data[1],"%")
```

```python
# Q10 - Select hours-per-week from dataframe where gender is female
data=df.loc[(df.gender=='Female'),['hours-per-week','gender']]
print(data)
print("average ", data.mean())

print("total Female enteries : ",len(data))
```

```python
# Q11 - Select hours-per-week from dataframe where gender is male

data=df.loc[(df.gender=='Male'),['hours-per-week','gender']]
print(data)
print("average ", data.mean())

print("total Male enteries : ",len(data))
```
